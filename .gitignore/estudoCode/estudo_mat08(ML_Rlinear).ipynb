```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
%matplotlib inline
```

```python
USA_houses = pd.read_csv("C:/Users/Gabriel/AulasUdemy/original/Python-Data-Science-and-Machine-Learning-Bootcamp/5. Machine Learning/Regressões Lineares/USA_Housing.csv")
```

```python
USA_houses                 
```

```python
X = USA_houses[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
Y = USA_houses['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.4, random_state=101)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
lm.intercept_              #intercepta o eixo y (x=0)
```

```python
lm.coef_                   # coeficientes gerados, um pra cada coluna do dataFrame 
```

```python
coeficientes = pd.DataFrame(lm.coef_, X.columns, columns=['coeficientes'])
coeficientes 

# Pra cada passo andado em X, o coeficiente é o crescimento relativo em Y
```

```python
predict = lm.predict(X_test)              # Aqui que ele faz as previsões dos preços(y) para cada índice do X_test
```

```python
plt.figure(figsize=(12,6))                # Vamos comparar então os valores gerados (predict) com os 
plt.scatter(Y_test, predict)              #                 valores reais dos mesmos índices(Y_test)
```

## Outras formas de avaliar:

```python
sea.distplot((Y_test-predict))
```

```python
from sklearn import metrics
```

```python
print('MAE',metrics.mean_absolute_error(Y_test,predict))                 
```

```python
print('MSE',metrics.mean_squared_error(Y_test,predict))
```

```python
print('RMSE',np.sqrt(metrics.mean_squared_error(Y_test,predict)))
```
