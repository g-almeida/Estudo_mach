```python
import seaborn as sea 
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sea.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
sea.distplot(tips['total_bill'],kde=False)
```

```python
sea.jointplot('total_bill','tip', data=tips, kind='reg')
```

```python
sea.pairplot(tips, hue = 'smoker' )
```

```python
np.array(tips.total_bill).std()
```

```python
flights = sea.load_dataset('flights')
flights
```
