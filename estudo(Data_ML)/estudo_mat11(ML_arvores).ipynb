# Árvores de Decisão e Florestas Aleatórias

- Similar ao de regressão logística

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/Gabriel/AulasUdemy/original/Python-Data-Science-and-Machine-Learning-Bootcamp/5. Machine Learning/Árvores de decisão e Florestas Aleatórias/kyphosis.csv')
```

```python
df
```

```python
sea.pairplot(df, hue='Kyphosis',diag_kind='hist')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis', axis = 1)
Y = df['Kyphosis']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3)
```

# Árvores de decisão

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,Y_train)
```

```python
pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,pred))
```

# Florestas Aleatórias

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 200)
```

```python
rfc.fit(X_train,Y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(Y_test,rfc_pred))
```
