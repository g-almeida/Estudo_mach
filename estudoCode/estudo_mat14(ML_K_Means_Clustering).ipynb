# K Means Clustering

- Tenta agrupar os dados em K grupos baseados em *características similares* de maneira NÃO SUPERVISIONADA
- A escolha do K não é trivial, as vezes convém utilizar o *método do cotovelo*
- Diagrama de Voroni
- Alguns exemplos de aplicações: cluster de docs, cluster de clientes, segmentação de mercado, geoestatística...

```python
from sklearn.datasets import make_blobs
```

```python
data= make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=50)
```

```python
plt.figure(figsize=(12,8))                                            # data[0] são os prórios dados gerados
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='jet')        # data[1] é a classe desses elementos
```

```python
data[0]
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)               # escolhemos 4 pois sabemos que existem 4 classes, basta analisar o data[1] acima
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(12,8))
ax1.set_title('KMeans')
ax1.scatter(data[0][:,0], data[0][:,1], c=km.labels_, cmap='rainbow')
ax2.set_title('Cluster original dos dados')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
