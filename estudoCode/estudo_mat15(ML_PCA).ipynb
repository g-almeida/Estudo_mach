# Análise do Componente Principal (PCA)

- Princípio similar à regressão linear, porém trabalha com um *conjunto ortogonal* para melhor descrição dos pontos
- NÃO SUPERVISIONADO
- Costuma exigir a *normalização dos dados*
- Usado apenas para análise
- Descarta parâmetros que praticamente só geram ruído. (Pontos muito fora da curva)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()                              # formato de um dicionário
type(cancer)
```

```python
cancer.keys()
```

```python
cancer['target_names']
```

```python
df = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
scaled_data.shape
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x_pca[:,0], x_pca[:,1], c= cancer['target'], cmap='plasma')
plt.xlabel('primeiro componente principal')
plt.ylabel('segundo componente principal')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,8))
sea.heatmap(df_comp, cmap='plasma')
```
