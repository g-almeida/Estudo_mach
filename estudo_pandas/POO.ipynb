## Métodos de classe:
    Métodos que se referem à própria classe e não a cada instância.

```python
class Pessoa:
    ano_atual = 2019
    
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        
    def get_ano_nascimento(self):
        print(self.ano_atual - self.idade)
    
    @classmethod
    def por_ano_nascimento(cls, nome, ano_nascimento):
        idade = cls.ano_atual - ano_nascimento
        prego = 'prego'
        return cls(nome, idade)
        
```

```python
pessoa = Pessoa('jones', 25)
pessoa.idade
```

```python
pessoa2 = Pessoa.por_ano_nascimento('joao', 1998)
pessoa2.idade
```

## Métodos Estáticos

```python
from random import randint

class PessoaME:
    ano_atual = 2019
    
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        
    def get_ano_nascimento(self):
        print(self.ano_atual - self.idade)
    
    @classmethod
    def por_ano_nascimento(cls, nome, ano_nascimento):
        idade = cls.ano_atual - ano_nascimento
        prego = 'prego'
        return cls(nome, idade)
        
    @staticmethod
    def geraID():
        rand = randint(10000, 19999)
        return rand            # Criando a ID
    
    def attID(self, ID):
        self._ID = ID          # Atribuindo a ID à instância
```

```python
pessoa3 = PessoaME('prego', 23)
pessoa3.geraID()
```

```python
pessoa3.attID(pessoa3.geraID())
pessoa3._ID
```

## Getters e Setters (?)

## Encapsulamento

```python
_dados = {}
_dados['clientela'] = {232 : 'brabo'}
```

```python
class BaseDeDados:
    def __init__(self):
        self.dados = {}
        
    def inserir_cliente(self, ID, nome):
        if 'clientes' not in self.dados:
            self.dados['clientes'] = {ID : nome}
        else:
            self.dados['clientes'].update({ID : nome})
    
    def lista_clientes(self):
        for cada in self.dados['clientes']:
            print (cada)
```

```python
bd = BaseDeDados()
bd.inserir_cliente(1, 'Otávio')
bd.inserir_cliente(2, 'Miranda')
bd.inserir_cliente(3, 'Rose')
```

```python
bd.dados
```

```python
bd.lista_clientes() 
```
